# A guide to using matplotlib for plotting data in python

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-100, 100, 201)

y = np.log(x) * 1000

```

```python
plt.plot(x, y, 'r') # a normal red line
plt.show()
```

```python
plt.plot(x, y, 'g.') # a green dotted line
plt.show()
```

```python
plt.plot(x, y, 'b--') # a blue dashed line
plt.show()
```

```python
plt.plot(x, y, 'c*') # a cyan line marked with asteriks
plt.show()
```

```python
plt.plot(x, y, 'ms') # a magenta line marked with squares
plt.show()
```

```python
plt.plot(x, y, 'y^') # a yellow line made with small triangles
plt.show()
```

### Those shapes are called markers, more marker options can be found at

## Making Subplots

```python
x = np.arange(1, 100, 0.2)
y1 = np.sin(x)
y2 = x ** 2 + 2 * x
y3 = np.log(x)
```

```python
ax1 = plt.subplot(221) # 2 rows 2 cols 1st plot
ax2 = plt.subplot(222) # 2 rows 2 cols 2nd plot
ax3 = plt.subplot(223) # 2 rows 2 cols 3rd plot
ax4 = plt.subplot(224) # 2 rows 2 cols 4th plot
```

```python
ax1 = plt.subplot(221)
ax1.plot(x, y1, 'r')
ax2 = plt.subplot(222)
ax2.plot(x, y2, 'y')
ax3 = plt.subplot(223)
ax3.plot(x, y3, 'b')
ax4 = plt.subplot(224)
ax4.plot(x, y1, 'g')
```

You can see in the above plot, that the yellow plot's Y co-ordinates are operlapping the red plot.
We can overcome this problem using the tight_layout() funtion.

```python
ax1 = plt.subplot(221)
ax1.plot(x, y1, 'r')
ax2 = plt.subplot(222)
ax2.plot(x, y2, 'y')
ax3 = plt.subplot(223)
ax3.plot(x, y3, 'b')
ax4 = plt.subplot(224)
ax4.plot(x, y1, 'g')

plt.tight_layout() # so that the plots do not overlap eachother
```

## Styling the plots

```python
from matplotlib import style
```

```python
print(plt.style.available) # this will show us what are available styles we can use
```

```python
x = np.arange(0, 30, 0.2)
y1 = np.sin(x)
y2 = np.cos(x)
```

Let's use some of those styles

```python
style.use('ggplot')
# let us mark which line indicates which data
plt.plot(x, y1, label="Sine")
plt.plot(x, y2, label="Cosine")
plt.legend(loc='upper left') # this decides the location of the label
plt.show()
```

```python
style.use('fivethirtyeight')
# let us mark which line indicates which data
plt.plot(x, y1, label="Sine")
plt.plot(x, y2, label="Cosine")
plt.legend(loc='upper left') # this decides the location of the label
plt.show()
```

```python
style.use('grayscale')
# let us mark which line indicates which data
plt.plot(x, y1, label="Sine")
plt.plot(x, y2, label="Cosine")
plt.legend(loc='upper left') # this decides the location of the label
plt.show()
```

```python
style.use('Solarize_Light2') # this style looks good
plt.grid(True) # For putting grids in plot
plt.plot(x, y1, label="Sine")
plt.plot(x, y2, label="Cosine")
plt.legend(loc='upper left') # this decides the location of the label
plt.show()
```

```python
plt.title("Sine Function") # For putting title
plt.suptitle("Styling Plots") # For putting super title
```

```python
plt.xlabel("No. of students") # For labeling X axis
plt.ylabel("No. of teachers") # For labeling Y axis
```

## Let's do some scatter plots

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', marker='2', s=100) # s is for size of the plots, markers can be found at above mentioned link

plt.show()
```

### Let's plot some histograms

```python
mu, sigma = 6, 2

x = mu + sigma * np.random.randn(10000)

plt.hist(x, 100, facecolor='blue', density=True)
plt.xlabel("Heights")
plt.ylabel("Percentage of students")
plt.title("Heights of students")
plt.text(-1, 0.20, "u = 6ft., sig = 2")
plt.grid(True)

plt.show()
```

## Let's play a little with the z axis

```python
ax = plt.axes(projection='3d')

z = np.linspace(0, 30, 100)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z)
```

```python
ax = plt.axes(projection='3d')
def z_fun(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = z_fun(X, Y)

ax.plot_surface(X, Y, Z)

plt.show()
```

## Let's plot bar graphs

```python
python = (50, 67, 69, 98)
js = (85, 26, 89, 37)
networking = (60, 20, 56, 22)
ml = (70, 23, 40, 89)

peoples = ["Rahul", "Sourav", "Debjeet", "Abhishek"]

index = np.arange(4)

plt.bar(index, python, width=0.2, label="Python")
plt.bar(index + 0.2, js, width=0.2, label="JavaScript")
plt.bar(index + 0.4, networking, width=0.2, label="Networking")
plt.bar(index + 0.6, ml, width=0.2, label="Machine Learning")

plt.title("IT Skill Levels")
plt.xlabel("Person")
plt.ylabel("Skill Level")
plt.xticks(index + 0.2, peoples)
plt.ylim(0, 150)
plt.legend(loc='upper right')

plt.show()
```

## Let's plot a pie chart

```python
nationalities = ["Indian", "Bangladeshi", "Pakistani", "Others"]

peoples = [63, 23, 21, 28]

pairs = zip(peoples, nationalities)

pairs = sorted(list(pairs))
peoples, nationalities = zip(*pairs) # * means all of pairs

# let's cut a pice of pie for you ❤
explode = [0, 0, 0.1, 0]

plt.pie(peoples, labels=nationalities, autopct='%.2f%%', shadow=False, explode=explode, counterclock=False, startangle=90)

plt.title("People in India")

plt.show()
```
