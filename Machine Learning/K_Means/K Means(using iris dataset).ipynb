```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
X, y = iris.data, iris.target
```

Lets, the find the number of cluster that is k

```python
k_rg = range(1, 10)
sse = []
for k in k_rg:
    kn = KMeans(n_clusters=k)
    kn.fit(X, y)
    sse.append(kn.inertia_)

sse
```

```python
plt.style.use(["dark_background"])
plt.rcParams["figure.figsize"] = (15, 5)

plt.plot(k_rg, sse, marker="o")
```

So the optimal cluster number is k=3

```python
km = KMeans(n_clusters=3) #Applying Kmeans classifier
y_pred = km.fit_predict(X)
y_pred
```

```python
km.cluster_centers_
```

```python
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s = 10, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s = 10, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s = 10, c = 'green', label = 'Iris-virginica')   #Visualising the clusters - On the first two columns

plt.scatter(km.cluster_centers_[:,   0], km.cluster_centers_[:,1], s = 100, marker="*",  c = '#fff', label = 'Centroids')   #plotting the centroids of the clusters

plt.legend()
```
